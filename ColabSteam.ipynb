```python


#@title 🌐 Colab Region Check  { display-mode: "form" }

import subprocess
import time
from google.colab import runtime

# Get region
region = subprocess.check_output(
    "curl -s ipinfo.io | grep region | cut -d\\\" -f4",
    shell=True
).decode().strip()

print(f"Your current region: {region}")

try:
    print("Automatically keeping this runtime in 15 seconds... (Press Ctrl+C to unassign)")
    for i in range(15, 0, -1):
        print(f"{i}...", end="", flush=True)
        time.sleep(1)
    print("\nKeeping current runtime...")

except KeyboardInterrupt:
    print("\nCtrl+C detected — unassigning runtime...")
    runtime.unassign()
```
